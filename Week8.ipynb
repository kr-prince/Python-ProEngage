```python
# Generate Random Password
# 1. Minimum 8 characters
# 2. Has to have a special character, numbers, caps, digits, and  small case
# Template : <special char><2 Capitals><3 Digits><3 small Letters>

from random import choice, choices
from string import ascii_uppercase, digits, ascii_lowercase

def generate_password():
    special_chars = '#$%&*@+'
    random_pass = choice(special_chars)+''.join(choices(ascii_uppercase, k=3))+\
                    ''.join(choices(digits, k=3))+''.join(choices(ascii_lowercase, k=3))
    return random_pass
```

```python
generate_password()
```

```python
# HangMan Game

# random list of fruits
fruits_name=['apple', 'banana', 'mango', 'guava', 'apricot', 'watermelon', 
             'strawberry', 'grapes', 'orange', 'pomegranate']
```

```python
from random import choice
```

```python
# number of chances to user
# max([len(set(fruit)) for fruit in fruits_name])
# max is 9 lets say we give 10 chances to user
chances = 10
```

```python
tot_score = 0
while True:
    fruit = choice(fruits_name)
    turn, score, correct_guess = 1, 0, []
    while turn <= chances:
        # take the char which user guessed
        guess = input("\nGuess your alphabet.. :  ")
        
        # check if user guessed a single character and alphabet
        if len(guess) > 1:
            print("You can only input a single character..")
            continue
        if not guess.isalpha():
            print("You can only input alphabets..")
            continue

        # check if the guess is correct or wrong
        guess = guess.lower()
        if guess in fruit:
            print("Correct guess..!")
            correct_guess.append(guess)
        else:
            print("Wrong..! You have {} chances left.".format(chances-turn))
            # increment turn taken
            turn += 1
        
        print("\n"+" ".join([fc if fc in correct_guess else '_' for fc in fruit]))

        # if the game is complete
        if set(fruit)==set(correct_guess):
            score = (chances-turn+1)*10
            print("Congrats..! You got the fruit. Your Score : {} ".format(score))
            tot_score += score 
            break

    carryOn = input("\nDo you want to play more?(y/n)  ")
    if carryOn.lower() not in ['y', 'yes']:
        print("\nYour total score was {}".format(tot_score))
        break
```
