https://www.hackerrank.com/challenges/text-wrap/problem

```python
# Using textwrap
import textwrap

def wrap(string, max_width):
    return '\n'.join(textwrap.wrap(string,max_width))

if __name__ == '__main__':
    string, max_width = input(), int(input())
    result = wrap(string, max_width)
    print(result)
```

```python
#Without using textwrap

def wrap(string, max_width):
    return '\n'.join([string[i:i+max_width] for i in range(0,len(string),max_width)])

if __name__ == '__main__':
    string, max_width = input(), int(input())
    result = wrap(string, max_width)
    print(result)
```

https://www.hackerrank.com/challenges/defaultdict-tutorial/problem

```python
from collections import defaultdict
n, m = map(int, input().strip().split())
dictA = defaultdict(list)
for i in range(n):
    dictA[input().strip()].append(i+1)
# print(dictA)
for i in range(m):
    b = input().strip()
    if b not in dictA:
        print("-1")
    else:
        print(" ".join(map(str, dictA[b])))
```

https://www.hackerrank.com/challenges/py-collections-namedtuple/problem

```python
from collections import namedtuple
n = int(input())
Student = namedtuple('Student', input().strip().split())
# print(Student)
total_marks = 0
for i in range(n):
    total_marks += int(Student(*input().strip().split()).MARKS)
print(round(total_marks/n, 2))
```

```python
# Shorter Version
from collections import namedtuple
n, Student = int(input()), namedtuple('Student', input().strip().split())
marks = [int(Student(*input().strip().split()).MARKS) for i in range(n)]
print(round(sum(marks)/n, 2))
```

https://www.hackerrank.com/challenges/py-collections-deque/problem

```python
from collections import deque
dq = deque()
for i in range(int(input())):
    inp = input().strip().split()
    if inp[0] == 'append':
        dq.append(inp[1])
    elif inp[0] == 'appendleft':
        dq.appendleft(inp[1])
    elif inp[0] == 'pop':
        dq.pop()
    elif inp[0] == 'popleft':
        dq.popleft()
    else:
        pass
print(" ".join(dq))
```

```python
# Shorter version - Using get attributes
from collections import deque
dq = deque()
for i in range(int(input())):
    inp = input().strip().split()
    getattr(dq, inp[0])(*[inp[1]] if len(inp) > 1 else [])
print(" ".join(dq))
```

```python
from collections import Counter
n = int(input().strip())
wordCount = Counter([input().strip() for i in range(n)])
# print(wordCount)
print(len(wordCount))
print(*wordCount.values())
```
