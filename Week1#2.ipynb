### Basic Data Types

A data type or simply type is an attribute of data which tells the compiler or interpreter how the programmer intends to use the data

Some of Python's standard data-types are as follows:

- Text Type:	str
- Numeric Types:	int, float, complex
- Sequence Types:	list, tuple
- Mapping Type:	dict
- Set Types:	set, frozenset
- Boolean Type:	bool

```python
# Number

# Python represents 3 different numerical data-types (int, float and complex). 
# By default all Python int can be arbitrarily high(depending on memory) so there is no
# separate concept of long integers  
# They can also be represented in Hexa-Decimal or Octal form


a=10
print(a)
print(type(a))
a=15.54
print(a)
print(type(a))
a = 100000000000000000000000000
print(a)
print(type(a))
c = complex(5)
print(c)
print(type(c))
c = complex(5, 7)
print(c)
print(type(c))
num = 0xA0F # Hexa-decimal
print(num)
print(type(num))
num = 0o37 # Octal
print(num)
print(type(num))
```

```python
# Boolean

# Booleans represent one of two values: True or False

a = True
print(a)
print(type(a))

# The bool() function allows you to evaluate any value, and give you True or False in return

print(bool(10<9))

# Any string is True, except empty strings like "" or ''.
# Any number is True, except 0.
# Any list, tuple, set, and dictionary are True, except empty ones like {},[]
```

```python
# String

# Like many other standard programming languages, strings in Python are arrays of 
# bytes representing unicode characters. Python does not have a character data type, 
# a single character is simply a string with a length of 1.

str1 = 'This is first string'
str2 = "This is second string"
str3 = """
Hey... This one is a really 
really big string, spread over
multiple lines.
"""
print(str1, type(str1))
print(str2, type(str2))
print(str3, type(str3))

print(str1[0])      # We can use string indexes like lists
print(len(str1))    # Len function for strings
print(str1[1:10])   # This is string slicing. [m:n] slices from index m to n-1
print(str1[-1:-10]) # -ve indexes is slicing from the back
print(str1+str2)    # String concatenation
print("This" in str1)# Membership

# String can also be formatted in Python like C 
print ("My name is %s and Roll No. is %d." % ('Prince', 11)) 
print ("My name is {} and Roll No. is {}.".format('Prince', 11))    # Using braces and format
print ("My name is {1} and Roll No. is {0}.".format(11, 'Prince'))  # Using braces,format and Index numbers
```

```python
# Strings are Immutable 
# Python does not allow direct update of Strings. You can create new or concatenate two other
# When a string reference is reinitialized with a new value, it creates a new object rather 
# than overwriting the previous value. In Python, strings are made immutable so that programmers 
# cannot alter the contents of the object (even by mistake). 

name = "Aarun" 
name[0] = 'T' 
name = "Tarun"
```

```python
# List

# Python lists are heterogenous collection of values of any types

list1 = [1,'A', 3.456, 'This is a string', '4th']         # Using square brackets
list1 = list((1,'A', 3.456, 'This is a string', '4th'))   # USing list constructor
print(list1, type(list1))
print(list1[2:4])                 # Slicing
print(list1[-3:-1])               # -ve indexes, slicing from back
list1[1] = "Index 1"              # Updating index 1
print(list1)
print(len(list1))                 # Len of list
print(list1+[3.14, 'Second item'])# List Concatenation
list1.append('last item')         # Append or add element to list
print(list1) 
print([1]*10)                     # Repeatition
print(1.23 in list1)              # Membership
list1.remove(1)                   # Removes some specific item
print(list1)
print(list1.pop())                # removes and returns the last item added
print(list1.pop(2))               # removes and returns the item at index 2
del list1[1]                      # deletes item at index 1
del list1                         # deletes the whole list
```

```python
# Tuple

# Python tuples are heterogenous collection of values of any types which
# CANNOT BE CHANGED

tuple1 = (1,'A', 3.456, 'This is a string', '4th')         # Using round brackets
print(tuple1, type(tuple1))
tuple1 = tuple((1,'A', 3.456, 'This is a string', '4th'))   # USing tuple constructor, can also pass list
print(tuple1[2:4])                 # Slicing
print(tuple1[-3:-1])               # -ve indexes, slicing from back
# tuple1[1] = "Index 1"              # ERROR : Tuple cannot be changed 
# print(tuple1)
print(len(tuple1))                 # Len of Tuple
print(tuple1+(3.14, 'Second item'))# Tuple Concatenation
print((1,)*10)                     # Repeatition
# For tuple with single value the correct way is (val,) NOT (val)
print(1.23 in tuple1)
# Cannot delete single tuple element but can delete the whole tuple
del tuple1                         # deletes the whole tuple
```

```python
# Set 

# Python Sets are heterogenous collection of values of any types which
# are UNORDERED, UNINDEXED and ARE UNIQUE values

set1 = {1,2,3,4,4,5,5,'abc'}        # Using curly braces
set1 = set((1,2,3,4,4,5,5,'abc'))   # Using set constructor, we can pass list or tuple
print(set1, type(set1))
print(len(set1))                    # Len of Set
# print(set1[1])                    # ERROR
set1.add('el')                      # add element to set
print(set1)
set1.update(['a','x','z'])          # add multiple elements, can also pass another set
print(set1)
set1.discard(4)                     # Removes element from set, does nothing if element not present
print(set1)
set1.remove(5)                      # Removes element from set, throws error if element not found
print(set1)
print(set1.pop())                   # Removes and returns random element
```

```python
# Some set operations

set1 = {1, 2, 3, 4, 5}
set2 = {3, 4, 5, 6, 7}

print(set1 | set2)           # Union operation
print(set1.union(set2))
print(set1 & set2)           # Intersection operation
print(set1.intersection(set2))
print(set1 - set2)           # Set difference
print(set1.difference(set2))
print(set1 ^ set2)           # Symmetric difference. ((A-B) U (B-A))
print(set1.symmetric_difference(set2))
```

```python
# Frozen Sets

# Frozen sets are just like sets but they CANNOT BE CHANGED or MODIFIED
# Tuples : Lists :: Frozensets : Sets

set1 = frozenset([1, 2, 3, 4])
set1.add(5)
```

```python
# Dictionary

# Dictionary is Python is like hash-map. It is used to store data in the form of 
# Key-Value pairs. 
# Dictionary is a collection of data which can be accessed by its Keys (hence 
# extremly fast access), are ordered (after Python 3.7 version), and do not allow 
# duplicate Keys

dict1 = {'k1':1, 'k2':2, 'k3':'Three'}                # Using curly braces
print(dict1, type(dict1))
print(dict1['k2'])                                    # Access using key
print(dict1.get('k2'))                                # USing get function
print(dict1.keys())                                   # All Keys
print(dict1.values())                                 # All values
dict1['k2'] = 'Two'                                   # Direct Update
print(dict1)
dict1.update({'k1': 'One'})                           # Update using another dict
print(dict1)
dict1['k4'] = 4                                       # Direct Add
print(dict1)
dict1.update({'k4': 4})                               # Add using another dict
print(dict1)
print(len(dict1))                                     # Len of dict
print(dict1.pop('k4'))                                # Removes Key and returns the value
del dict1['k3']                                       # Deletes the particular Key-Val pair
print(dict1)
dict1.clear()                                         # Clears the whole dict, but it exists
print(dict1)
del dict1                                             # Deletes the whole dict, does not exists now
```
