```python
# Using for loop

# input reads the key board input and returns it as string
num=int(input())
for n in range(1, num+1):  # range sequence, be default starts from 0
    print(n,end='')

# Using while loop
num=int(input())
n=1
while n <= num:
    print(n, end='')
    n += 1

# List Join
# range returns the seq
num=int(input())
# map applies some function to each elemnet of some sequence and returns iterator
''.join(map(str, list(range(1, num+1))))
```

https://www.hackerrank.com/challenges/python-string-split-and-join/problem

```python
def split_and_join(line):
    # split method of string and join method for list
    return '-'.join(line.split(' '))
    
if __name__ == '__main__':
    line = input()
    result = split_and_join(line)
    print(result)
```

https://www.hackerrank.com/challenges/python-lists/problem

```python
if __name__ == '__main__':
    N = int(input())
    myList = []
    for i in range(N):
        command = input()
        # startswith method of string
        if command.startswith('insert'):
            myList.insert(int(command.split(' ')[1]), int(command.split(' ')[2]))
        elif command.startswith('print'):
            print(myList)
        elif command.startswith('remove'):
            myList.remove(int(command.split(' ')[1]))
        elif command.startswith('append'):
            myList.append(int(command.split(' ')[1]))
        elif command.startswith('sort'):
            myList.sort()
        elif command.startswith('pop'):
            myList.pop()
        elif command.startswith('reverse'):
            myList.reverse()
        else:
            pass
```

https://www.hackerrank.com/challenges/py-introduction-to-sets/problem

```python
def average(array):
    distinctNums = set(array)
    average = sum(distinctNums)/len(distinctNums)
    return average

if __name__ == '__main__':
    n = int(input())
    arr = list(map(int, input().split()))
    result = average(arr)
    print(result)
```
