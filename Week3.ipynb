### Operators

Python language supports the following types of operators âˆ’

- Arithmetic Operators
- Comparison (Relational) Operators
- Assignment Operators
- Logical Operators
- Bitwise Operators
- Membership Operators
- Identity Operators

```python
# Arithmetic Operators : For simple mathematical operations like
# addition, subtraction, multiplication etc

x, y = 10, 21

print(x + y)    # Add
print(y-x)      # Substract
print(x*y)      # Multiply
print(y/x)      # Division
print(y%x)      # Mod division - Remainder
print(y//x)     # Floor division - Intger val of quotient
print(y**x)     # Exponent(power)
```

```python
# Comparison Operators (also called Relational Operator)
# Compare the values on left and Right and return boolean output if the relation holds

x, y = 10, 21

print(x < y)       # Less than 
print(x > y)       # Greater than
print(x >= y)      # Greater than equals to
print(x == y)      # Is Equal to, NOT TO BE CONFUSED WITH =
print(x != y)      # Not equal to
```

```python
# Assignment Operators
# Used to assign values to variables

x = 5
print(x)
x += 12     # Add and then assign 
print(x)
x /= 2      # Divide and then assign
print(x)
```

```python
# Logical Operators
# Used to combine and evaluate conditional statements
# AND, OR, NOT

print(10>9 and 1<0)  # (True and False) => False
print(10>9 or 1<0)   # (True or True) => True
print(not(10<9))
```

```python
# Bitwise Operators
# Used in bit-by-bit operations

# Operator	        Name	           Description
#   & 	             AND	           Sets each bit to 1 if both bits are 1
#   |	              OR	           Sets each bit to 1 if one of two bits is 1
#   ^	             XOR	           Sets each bit to 1 if only one of two bits is 1
#   ~ 	             NOT	           Inverts all the bits, (num in 2's complement form due to a 
#                                      signed binary number)
#   <<	     Zero fill left shift	   Shift left by pushing zeros in from the right and let the 
#                                      leftmost bits fall off
#   >>	      Signed right shift	   Shift right by pushing copies of the leftmost bit in from 
#                                      the left, and let the rightmost bits fall off
```

```python
a = 60            # 60 = 0011 1100 
b = 13            # 13 = 0000 1101 
c = 0

c = a & b;        # 12 = 0000 1100
print("Line 1 - Value of c is ", c)

c = a | b;        # 61 = 0011 1101 
print("Line 2 - Value of c is ", c)

c = a ^ b;        # 49 = 0011 0001
print("Line 3 - Value of c is ", c)

                  # 0011 1100 => 1100 0011
c = ~a;           # -61 = 1100 0011
print("Line 4 - Value of c is ", c)

c = a << 2;       # 240 = 1111 0000
print("Line 5 - Value of c is ", c)

c = a >> 2;       # 15 = 0000 1111
print("Line 6 - Value of c is ", c)
```

```python
# Membership Operators 
# Used to test if a sequence is present in an object or not

# IN or NOT IN
list1 = ['a', 'b', 10]
print(11 not in list1)
```

```python
# Identity Operators 
# Used to compare objects, not if they are equal, but if they are actually  
# the same object, with the same memory location

# is , is not
a = "Prince"
b = "Prince"
c = "Hello"
print(a is b)
print(a is c)
```

### Decision Making

Decision making helps test and evaluate certain conditions in the program and based on the results execute some specific set of instructions 
- Any non-zero and non-null values are evaluated as TRUE 
- Any zero or null values are evaluated as FALSE

```python
# Simple If
var1 = 0
if var1:
   print("Hey this is True")
   print(var1)
print("End")
```

```python
# If-Else
var1 = 100
if var1:
   print("Hey this is True")
   print(var1)
else:
   print("Naah. Its False")
   print(var1)
print("End")
```

```python
# If-Elif-Else
marks = 65
if marks>80:
   print("Grade A")
   print(marks)
elif marks>60:
   print("Grade B")
   print(marks)
elif marks>40:
   print("Grade C")
   print(marks)
else:
   print("Grade F")
   print(marks)
print("End")
```

```python
# Nested If-Else 
num = int(input("enter number"))
if num%2 == 0:
   if num%3 == 0:
      print ("Divisible by 3 and 2")
   else:
      print ("divisible by 2 not divisible by 3")
else:
   if num%3 == 0:
      print ("divisible by 3 not divisible by 2")
   else:
      print  ("not Divisible by 2 not divisible by 3")
```

### Control Flow - Loops

In general, like any standard programming language the statements in Python are excuted sequentially, but sometimes some specific few lines of code need to be executed multiple times.

Python has two main types of loops
- for loop
- while loop

Loop control statements : Loop control statements change the execution from its normal sequence
- Break : Terminates the loop statement and transfers execution to the statement immediately following the loop
- Continue : Causes the loop to skip the remainder of its body and immediately retest its condition prior to reiterating
- Pass : The pass statement in Python is used when a statement is required syntactically but you do not want any command or code to execute


```python
# For loop

# for iterating_var in sequence:
#    statements(s)

fruits = ['banana', 'apple',  'mango']
# for fruitname in fruits:
#     print("I love ", fruitname)

for index in range(len(fruits)):
   print ('Current fruit :', fruits[index])

print ("End")
```

```python
# Else block in loop is executed only if for loops terminates normally 
# (and not by encountering break statement)

numbers = [11,33,55,39,55,75,37,21,23,41,13]

for num in numbers:
   # print(num)
   if num%2 == 0:
      print ('the list contains an even number')
      break
else:
   print ('the list does not contains even number')
print("End")
```

```python
# while loop

# while expression:
#    statement(s)

count = 0
while count < 10:
   if count > 5:
       count = count + 1
       continue
   print(count, " is  less than 10")
   count = count + 1
else:
   print (count, " is not less than 10")
print("End")
```
